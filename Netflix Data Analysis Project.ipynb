### Q2. In which year highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
# Checking Unique Values in Year

netflix_df['Year'].unique()
```

```python
# Converting Year to 'int' Datatype

netflix_df['Year'].astype(int)
```

```python
# Generating Barplot

plt.figure(figsize=(10,6))
sns.countplot(x=netflix_df['Year'],hue=netflix_df['Category'])
plt.xlabel('Years', fontsize = 20)
plt.ylabel('Count of Categories', fontsize = 20)
plt.title('No. of Movies & TV Shows per Year', fontsize = 24, fontweight = 'bold')
plt.legend(loc='upper left')

plt.show()
```

### Q10. What is the maximum duration of a Movie/Show on Netflix ?

```python
# Creating a seprate Dataframe for movie records

movie_df = netflix_df[netflix_df['Category'] == 'Movie']
```

```python
#Splitting the Duration into new Column 'Run Time'

movie_df['Run Time'] = movie_df['Duration'].apply( lambda x:x.split(' ')[0])
```

```python
# Converting the datatype of Run Time Column to 'int'

movie_df['Run Time'] = movie_df['Run Time'].astype('int')
```

```python
# Top 5 Movie with highest RunTime

top5_movies = movie_df[['Title', 'Run Time']].sort_values(by='Run Time', ascending=False).head()
top5_movies
```

```python
# Generating Barplot

plt.figure(figsize=(10,6))
sns.barplot(x=top5_movies['Title'], y=top5_movies['Run Time'])
plt.xlabel('Movies', fontsize = 20)
plt.ylabel('Run Time', fontsize = 20)
plt.title('Top 5 Movies by Run Time', fontsize = 24, fontweight = 'bold')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Creating a seprate Dataframe for movie records

tvshow_df = netflix_df[netflix_df['Category'] == 'TV Show']
```

```python
#Splitting the Duration into new Column 'Season'

tvshow_df['Seasons'] = tvshow_df['Duration'].apply( lambda x:x.split(' ')[0])
```

```python
# Converting the datatype of Seasons Column to 'int'

tvshow_df['Seasons'] = tvshow_df['Seasons'].astype('int')
```

```python
# Top 5 Movie with highest RunTime

top5_shows = tvshow_df[['Title', 'Seasons']].sort_values(by='Seasons', ascending=False).head()
top5_shows
```

```python
# Generating a barplot

plt.figure(figsize=(10,6))
sns.barplot(x=top5_shows['Title'], y=top5_shows['Seasons'])
plt.xlabel('TV Shows', fontsize = 20)
plt.ylabel('Seasons', fontsize = 20)
plt.title('Top 5 TV Shows by Seasons', fontsize = 24, fontweight = 'bold')
plt.xticks(rotation = 45)

plt.show()
```
